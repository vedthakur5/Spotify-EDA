```python
tfa = df.groupby("artist").count().sort_values(by = "song_title", ascending = False)["song_title"][:5]
tfa
```

```python
tfa.plot.barh()
plt.show()
```

```python
tflt = df[["loudness", "song_title"]].sort_values(by = "loudness")[:5]
tflt
```

```python
plt.figure(figsize = (12,7))
sns.barplot(x = "loudness", y ="song_title", data = tflt)
plt.title("Top 5 loudest tracks")
plt.show()
```

```python
tfads = df[["danceability", "song_title", "artist"]].sort_values(by="danceability", ascending = False)[:5]
tfads
```

```python
plt.figure(figsize = (12,7))
sns.barplot(x = "danceability", y ="artist", data = tfads)
plt.title("Top 5 Artists with most danceability songs")
plt.show()
```

```python
tfit = df[["instrumentalness", "song_title", "artist"]].sort_values(by="instrumentalness", ascending = False)[:5]
tfit
```

```python
plt.figure(figsize =(12,7))
plt.pie(x = "instrumentalness", data = tfit, autopct = '%1.2f%%', labels = tfit.song_title)
plt.show()
```

```python
ifc = ['acousticness', 'danceability', 'duration_ms', 'energy',
       'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']
```

```python
for fc in ifc:
  pd = df[df["target"] == 1][fc]
  nd = df[df["target"] == 0][fc]

  plt.figure(figsize =(12,7))

  sns.distplot(pd, bins = 30, label = 'Positive', color = 'green')
  sns.distplot(nd, bins = 30, label = 'Negative', color = 'red')

  plt.legend("upper right")
  plt.title(f"Positive and Negative Histogram Plot for {fc}")
  plt.show()
```

```python
sns.regplot(x="energy", y="loudness", data = df)

# Set plot title and labels
plt.title("Scatter Plot between loudness and energy")
plt.xlabel("energy")
plt.ylabel("loudness")

# Show the plot
plt.show()
```

```python
sns.regplot(x="loudness", y="tempo", data = df)

# Set plot title and labels
plt.title("Scatter Plot between loudness and eenergy")
plt.xlabel("loudness")
plt.ylabel("tempo")

# Show the plot
plt.show()
```

```python
# Create line plot
sns.lineplot(x="liveness", y="loudness", data = df)

# Set plot title and labels
plt.title("Line Plot")
plt.xlabel("liveness")
plt.ylabel("loudness")

# Show the plot
plt.show()
```
